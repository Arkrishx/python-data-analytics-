```python
#pie chart

import matplotlib.pyplot as plt

food = ["Dosa","Idly","Poratta","Briyani"]
cost = [50,15,20,120]

plt.title("Food cost")
plt.pie(cost,labels=food)
plt.show()
```

```python
#pie chart with percentage

import matplotlib.pyplot as plt

food = ["Dosa","Idly","Poratta","Briyani"]
cost = [50,15,20,120]

plt.title("Food cost")
plt.pie(cost,labels=food,autopct="%0.1f%%", colors = ['Red','Green','Blue','Pink'])
plt.show()
```

```python
#monthly sales dashboard -> combines multiple charts and using subplot()

import matplotlib.pyplot as plt

month = ['Jan','Feb','Mar','Apr','May','Jun']
sales = [100000,120000,8000,16000,100000,6000]

# Create a single figure with a 2x2 grid of subplots
plt.figure(figsize=(15,10)) # Adjust figure size for a 2x2 layout

# Line chart (subplot 1, top-left)
plt.subplot(2,2,1)
plt.plot(month,sales)
plt.title("Monthly sales (Line Chart)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True) # Add grid for better readability

# Bar chart (subplot 2, top-right)
plt.subplot(2,2,2)
plt.bar(month,sales)
plt.title("Monthly sales (Bar Chart)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=90, ha='right') # Rotate labels to prevent collision
plt.grid(axis='y') # Add horizontal grid for better readability

# Pie chart (subplot 3, bottom-left)
plt.subplot(2,2,3)
# To address percentage collision, reduce fontsize of autopct labels
plt.pie(sales,labels=month,autopct="%0.1f%%", startangle=90, textprops={'fontsize': 8})
plt.title("Monthly sales (Pie Chart)")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

# Adjust layout to prevent overlapping elements
plt.tight_layout()
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7,8]
marks = [30,40,50,60,70,80,90,95]

plt.scatter(hours,marks)
plt.title("Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
