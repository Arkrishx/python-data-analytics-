# Handling duplicates


```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Alice'],
    'Age': [25, 30, 25],
    }

df = pd.DataFrame(data)

print(df.duplicated())
```

```python
print(df.drop_duplicates())
```

```python
#renaming columns

import pandas as pd
data = {
    "emp_nm":["john","mary","raj"],
    "emp_sal":[10000,20000,12000]
}
df = pd.DataFrame(data)

df.rename(columns={
    "emp_nm":"Employee_Name",
    "emp_sal":"Employee_Salary"
},inplace=True)

print(df)
```

```python
import pandas as pd

data = {
    "Region" : ["North","South","East","West","North","South"],
    "Product":["LAP","MOBILE","LAP","TAB","MOBILE","TAB"],
    "Sales":[10000,20000,15000,30000,25000,35000]
}

df = pd.DataFrame(data)
print(df)

south_sales = df[df["Region"]=="South"]
print("\nSouth Region Sales:\n",south_sales)

#product_sales = df.groupby("Product")["Sales"].sum()
#print("\nProduct-wise Sales:\n",product_sales)

high_sales = df[df["Sales"]>15000]
print("\nProducts with sales greater than 15000:\n",high_sales)
```

```python
#sorting

sorted_df = df.sort_values("Sales") #for ascending
print(sorted_df)

dec_sort = df.sort_values(by="Sales",ascending=False)
print("\n",dec_sort)
```
