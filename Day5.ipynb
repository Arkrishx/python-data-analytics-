# Remining of Day4

```python
#std() --> Standard deviation

import numpy as np
a = np.array([1,2,3,4,5])
print(np.std(a))
```

```python
# var() --> Variance

import numpy as np
a = np.array([1,2,3,4,5])
print(np.var(a))
```

```python
# min() & max()

import numpy as np
a = np.array([1,2,3,4,5])
print(np.min(a))
print(np.max(a))
```

```python
# argmin() & argmax()  --> index of min & max

import numpy as np
a = np.array([1,2,3,4,5])
print(np.argmin(a))
print(np.argmax(a))
```

```python
# cumsum() --> running sum or cumulative sum

import numpy as np
a = np.array([1,2,3,4,5])
print(np.cumsum(a))
```
