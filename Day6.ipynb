# Practice


```python
# Employee dataset exploration

import pandas as pd

emp = {
    "emp_id" : [111,112,113,114,115],
    "Name": ["Ak","raj","sam","tom","tyler"],
    "Dept": ["IT","HR","CS","EE","IT"],
    "Salary":[15000,10000,13000,18000,8000],
    "Exp":[2,3,5,7,1]
}


df = pd.DataFrame(emp)
print("Employees Dataset:")
print(df)


avg_salary = df['Salary'].mean()
print("\nAverage Salary:",avg_salary)

max_salary = df['Salary'].max()
print("\nMaximum Salary:",max_salary)

min_salary = df['Salary'].min()
print("\nMinimum Salary:",min_salary)

it_dept = df[df['Dept']=='IT']
print("\nEmployees in IT:\n", it_dept)

exp_emp = df[df['Exp']>2]
print("\nExperienced Employees:\n",exp_emp)

dept_count = df["Dept"].value_counts()
print("\nEmployee count by Dept:\n",dept_count)
```

# fillna() - function

```python
import pandas as pd

data = {
    "Name": ["Ak","raj","sam","tom","tyler"],
    "Age":[18,None,19,None,17]
}

df = pd.DataFrame(data)

print(df['Age'].fillna(0))
```

```python
import pandas as pd

data = {
    "Name": ["Ak","raj","sam","tom","tyler"],
    "Age":[18,None,19,None,17]
}

df = pd.DataFrame(data)

mean_age = df['Age'].mean()
print(mean_age)

df['Age'].fillna(mean_age,inplace=True)
print(df)
```

```python
#isnull() --> show the column data that is null or not. using TRUE/FALSE

#isnull().sum()--> sum the TRUE of a column and display that as a number.

#dropna()--> clean the missing values // remove the missing values and display the correct value only.

#fillna(0)--> display the missing value and fill that with 0.00 (zeros).
```
