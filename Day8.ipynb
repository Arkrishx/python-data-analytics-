```python
#group by concept

import pandas as pd

data = {
    "Region" : ["North","South","East","West","North","South"],
    "Product":["LAP","MOBILE","LAP","TAB","MOBILE","TAB"],
    "Sales":[10000,20000,15000,30000,25000,35000]
}

df = pd.DataFrame(data)
print(df)
```

```python
region_sales = df.groupby("Region")["Sales"].sum() #day8
print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
print(product_sales)
```

```python
#Aggregation concept

print(df["Sales"].sum())
```

```python
print(df["Sales"].max())
```

```python
print(df["Sales"].min())
```

```python
print(df["Sales"].count())
```

```python
#Groped data -> mathematical operation -> Aggregation

result = df.groupby("Region")["Sales"].agg(["sum","max","min","count"])
print(result)
```

# Data visualisation Basics

```python
#Line chart

import matplotlib.pyplot as plt
import pandas as pd

months = ["Jan","Feb","Mar","Apr"]
sales = [12000,18000,6000,22000]
plt.plot(months,sales)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

months = ["Jan","Feb","Mar","Apr"]
sales = [12000,18000,6000,22000]
plt.plot(months,sales,color='Green',marker=0,linestyle="--")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```

```python
#Bar chart

import matplotlib.pyplot as plt


months = ["Jan","Feb","Mar","Apr"]
sales = [12000,18000,6000,22000]
plt.bar(months,sales)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```

```python
import matplotlib.pyplot as plt


months = ["Jan","Feb","Mar","Apr"]
sales = [12000,18000,6000,22000]
plt.bar(months,sales,color='Red')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```
